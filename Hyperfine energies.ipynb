$ c = \lambda f $ becomes $\lambda = c/f$ so $\delta \lambda = \frac{c}{f^2}\delta f $

```python
def get_w_diff(df):
    c = 2.998e8 # m/s
    f_0 = 384.2304844685e12 # Hz (difference between 5_2_s_1/2 and 5_2_p_3/2 state)
    d_wavelength = c/(f_0**2) * df * 10**9 # nm
    # alternatively, d_wavelength = (c/f_0 - c/(f_0 + df)) * 10**9
    return d_wavelength
```

```python
# Frequency differences (Hz)
f_0 = 384.2304844685e12

f_p_f0 = 302.0738e6 # difference from F = 0 to 5_2_P_3/2 state
f_p_f1 = 229.8518e6 # difference from F = 1 to 5_2_P_3/2 state
f_p_f2 = 72.911232e6 # difference from F = 2 to 5_2_P_3/2 state
f_p_f3 = 193.7407e6 # difference from F = 3 to 5_2_P_3/2 state

f_s_f1 = 4.271676631815181e9 # difference from F = 1 to 5_2_s_1/2 state
f_s_f2 = 2.563005979089109e9 # difference from F = 2 to 5_2_s_1/2 state

freq = {
    '10': f_0-f_p_f0+f_s_f1, # from F = 1 to F = 0 (s to p state), etc.
    '11': f_0-f_p_f1+f_s_f1,
    '12': f_0-f_p_f2+f_s_f1,
    '21': f_0-f_p_f1-f_s_f2,
    '22': f_0-f_p_f2-f_s_f2,
    '23': f_0+f_p_f3-f_s_f2
}

wavelength = {}
for i in ['10', '11', '12', '21', '22', '23']:
    wavelength[i] = get_w_diff(freq[i])
    print(i, ' (nm): ', get_w_diff(freq[i]))

print('11-10: ', wavelength['11'] - wavelength['10'])
print('12-10: ', wavelength['12'] - wavelength['10'])
print('21-10: ', wavelength['21'] - wavelength['10'])
print('22-10: ', wavelength['22'] - wavelength['10'])
print('23-10: ', wavelength['23'] - wavelength['10'])

print('12-11: ', wavelength['12'] - wavelength['11'])
print('21-11: ', wavelength['21'] - wavelength['11'])
print('22-11: ', wavelength['22'] - wavelength['11'])
print('23-11: ', wavelength['23'] - wavelength['11'])

print('21-12: ', wavelength['21'] - wavelength['12'])
print('22-12: ', wavelength['22'] - wavelength['12'])
print('23-12: ', wavelength['23'] - wavelength['12'])

print('22-21: ', wavelength['22'] - wavelength['21'])
print('23-21: ', wavelength['23'] - wavelength['21'])

print('23-22: ', wavelength['23'] - wavelength['22'])

print('Average relative diff between small and big dip (nm)', np.average([0.01293544, 0.01289932, 0.012839, 0.01272712])) # from data eyeballing
print('Splitting between F=1/F=0 and F=2/F=3 in wavelength (nm): ', 780.2688996432626 - 780.2560272379031) # difference between the F=1 -> F=0 and F=2->3 transition
print('Splitting freq between F=1s/F=0p and F=2s/F=3p (GHz) (based on Rb transitions): ', (freq['10']-freq['23'])*10**-9, )

c = 2.998e8 # m/s
w_0 = 780.241209686*10**-9 # m (central wavelength)
print('Splitting freq between F=1s/F=0p and F=2s/F=3p (GHz) (based on data eyeballing): ', c/w_0**2 * 0.01286022e-18)
# multiply by average relative diff using main s1/2 to p3/2 transition
```
