```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
v = np.linspace(0, 50, 1000)
T=100
P = v**2*np.exp(-v**2/T)
plt.plot(v, P)
plt.fill_between(v, P, where=(v <= 20), color='blue', alpha=0.3)
plt.grid()
plt.xlabel('Speed of atom')
plt.ylabel('Population of atoms')
plt.title('Maxwell-Boltzmann Distribution of Atom Speeds')
plt.vlines([20], 0, 40, colors='r', linestyles='dashed')
plt.text(20.5, 32, r'$v_c$', size=15)
plt.ylim([-1, 40])
plt.savefig('capture_velocity.png', dpi=300)
```
